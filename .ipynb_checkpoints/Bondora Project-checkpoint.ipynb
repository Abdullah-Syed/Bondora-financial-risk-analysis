#### On-hot encoding for catagorical variable 

```python
#Import necessary package 
from sklearn.preprocessing import OneHotEncoder
```

```python
dataset_Bondora["Status"].unique()
```

```python
onehotencoder = OneHotEncoder()

transformed_data = onehotencoder.fit_transform(dataset_Bondora[["Status"]]).toarray()

# Now above transformed_data is an array 
# so we are converting it to a dataframe
encoded_data = pd.DataFrame(transformed_data, index=dataset_Bondora.index)

# now concatenate the original data and the encoded data using pandas
concatenated_data = pd.concat([dataset_Bondora, encoded_data], axis=1)
concatenated_data
```
